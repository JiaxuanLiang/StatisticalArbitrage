## Conclusion

In the Black-Scholes framework of stock price dynamics, i.e. 

$$S_t=S_0 e^{(\alpha - \frac{\sigma^2}{2})t + \sigma W_t}$$ 
$$r_f \text{: the risk-free rate}$$ 

In the Black-Scholes model, there exists statistical arbitrage if $\alpha− r_f \neq \frac{\sigma^2}{2}$. 
- If $\alpha−r_f > \frac{\sigma^2}{2}$, then there exists statistical arbitrage for the long until barrier strategies.
- If $\alpha−r_f < \frac{\sigma^2}{2}$, then there exists statistical arbitrage for the short until barrier strategies.

to exclude statistical arbitrage opportunities, the Sharpe ratio of any given stock must be bounded by $\frac{\sigma}{2}$. 

In a market where short sell is not allowed, the no-arbitrage condition is given by $0<\alpha−r_f<\frac{\sigma^2}{2}$.

If an investor knows or believes that he knows the stocks that satisfy the statistical arbitrage condition, then this is sufficient to design statistical arbitrage trading strategy. 


## Setting

### Statistical Arbitrage

#### Definition
Given the stochastic process for the discounted cumulative trading profits, denoted as $\{v(t): t \geq 0\}$ that is defined on a probability space $(\mathcal \Omega,\mathcal F,\mathcal P)$ the statistical arbitrage is defined as follows:

*A statistical arbitrage is a zero initial cost, self-financing trading strategy $\{v(t): t \geq 0\}$ with cumulative discounted value $v(t)$ such that*
- $v(0) = 0$
- $\lim_{t \to \infty}\mathbb E[v(t)]>0$
- $\lim_{t \to \infty} P(v(t)<0)=0$
- $\lim_{t \to \infty} \frac{var(v(t))}{t}=0$ if $P(v(t)<0)>0$, $\forall t<\infty$ 
    
#### Properties

**Property 1**

Given the probability space $(\mathcal \Omega,\mathcal F,\mathcal P)$ and stochastic process $\{v(t): t \geq 0\}$ defined on this space. Consider that we have the following properties for the trading strategy $\{v(t): t \geq 0\}$
- $v(0) = 0$
- $\lim_{t \to \infty} \mathbb E[v(t)]>0$
- $\lim_{t \to \infty} var(v(t))=0$ 

then conditions 1 − 3 implies $\lim_{t \to \infty} P(v(t)<0)=0$.

(Proof: consider $P(v(t)\leq\mu_t-\alpha)\leq \frac{\sigma_t^2}{\sigma_t^2+\alpha^2}$ by setting $\alpha=\mu_t$, where $\alpha \geq 0$, $\mu_t$ and $\sigma_t^2$ is the mean and variance of $v(t)$)

**Property 2**

Given the probability space $(\mathcal \Omega,\mathcal F,\mathcal P)$ and stochastic process $\{v(t): t \geq 0\}$ defined on this space. Consider that we have the following properties for the trading strategy $\{v(t): t \geq 0\}$
- $v(0) = 0$
- $\lim_{t \to \infty}\mathbb E[v(t)] = \infty$
- $\lim_{t \to \infty} var(v(t))=c$ 

then conditions 1 − 3 implies $\lim_{t \to \infty} P(v(t)<0)=0$.

(Proof: consider $P(v(t)\leq\mu_t-\alpha)\leq \frac{\sigma_t^2}{\sigma_t^2+\alpha^2}$ by setting $\alpha=\mu_t$ where $\alpha \geq 0$, $\mu_t$ and $\sigma_t^2$ is the mean and variance of $v(t)$)

**Property 3**

Given any two trading strategies $v_1(t)$, $v_2(t) \in \mathcal C$, their linear combination,
$v^∗(t) = a v_1(t) + (1 − a)v_2$, is also in $\mathcal C$.

(Proof: let $v^*=\alpha v_1+(1-\alpha)v_2$, where there is statistical arbitrage opportunity in $v_1$ and $v_2$, $v^*$ satisfies the definition as well)

We can minimize the variance of the linear combination of two statistical arbitrage strategies as $$\min_{a} a^2\sigma_1^2 + (1 − a)^2\sigma_2^2 + 2a(1 − a)\rho\sigma_1\sigma_2,$$ where $a \in [0,1]$, $\sigma_1$ and $\sigma_2$ is the standard deviation of $v_1 (t)$ and $v_2 (t)$, respectively. The optimal portfolio weights that minimize the variance are given by $$\hat a = \frac{\sigma_2^2-\rho\sigma_1\sigma_2}{\sigma_1^2+\sigma_2^2-2\rho\sigma_1\sigma_2}$$ $$1 - \hat a = \frac{\sigma_1^2-\rho\sigma_1\sigma_2}{\sigma_1^2+\sigma_2^2-2\rho\sigma_1\sigma_2}$$ for time $t$.


### Strategies

**Short Strategy**: a self-financing trading strategy by short selling the stock and investing
in the money market account at time 0, and renewing the short position and re-investing profits (or re-financing losses) in the money market account.

$$v(t) = S_0 \left(1-e^{(\alpha− r_f-\frac{\sigma^2}{2})t + \sigma W_t}\right)$$ 
$$\mathbb E[v(t)] = S_0\left(1-e^{(\alpha−r_f)t}\right) $$
$$var(v(t)) = \left(e^{\sigma^2 t}−1\right)S_0^2 e^{2(\alpha-r_f)t}$$

Statistical arbitrage opportunity exists when $\alpha−r_f \leq −\frac{\sigma^2}{2}$.

Proof:

When $\alpha−r_f \leq −\frac{\sigma^2}{2}$, there are:

$$\lim_{t \to \infty} \mathbb E[v(t)] = S_0 > 0 \text{ for } \alpha - r_f < 0$$
$$\lim_{t \to \infty} P(v(t)<0) = 0 \text{ for } \alpha - r_f < 0 $$
$$\lim_{t \to \infty} \frac{var(v(t))}{t} = 0 \text{ for } \alpha−r_f \leq −\frac{\sigma^2}{2}$$

**Buy & Hold Strategy**: a self-financing trading strategy that consists of buying and holding one unit of stock financed by the money market account, and re-investing profits (or re-financing losses) in the money market account.

$$v(t) = S_0 \left(e^{(\alpha− r_f-\frac{\sigma^2}{2})t + \sigma W_t}-1\right)$$
$$\mathbb E[v(t)] = S_0\left(e^{(\alpha−r_f)t}-1\right)$$
$$var(v(t)) = \left(e^{\sigma^2 t}−1\right)S_0^2 e^{2(\alpha-r_f)t}$$

Without the fourth condition, buy and hold strategies yield statistical arbitrage opportunities in the Black-Scholes model for $\alpha−r_f > \frac{\sigma^2}{2}$.

Proof:

When $\alpha−r_f > \frac{\sigma^2}{2}$, there are:

$$\lim_{t \to \infty} \mathbb E[v(t)] = \lim_{t \to \infty} S_0\left(e^{(\alpha−r_f)t}-1\right) > 0$$
$$\lim_{t \to \infty} P(v(t)<0) = 0$$
$$\text{ (violation) }\lim_{t \to \infty} \frac{var(v(t))}{t} = \infty \text{ for } \alpha−r_f > \frac{\sigma^2}{2}$$ 
$$(\text{since } \lim_{t \to \infty} \frac{var(v(t))}{t} = \infty \text{ for } \alpha−r_f > −\frac{\sigma^2}{2})$$ 


**Buy & Hold Until Constant Barrier Strategy**: a self-financing trading strategy with a termination condition that consists of buying and holding one unit of stock financed by the money market account with constant risk-free rate. Whenever the stock price process hits to the constant barrier level, sell the stock and invest in the money market account.

$$v_t = \left\{
\begin{array} \
Be^{-r_f t^*} -S_0, \text{ if } t^* \leq t & \\
S_te^{-r_f t} -S_0, \text{ else } & \\
\end{array} \right. 
$$ where $t^* = min\{t \geq 0: S_t = B\}$ and $B>S_0$ is the constant barrier level

Define $X_t = \frac{ln \left(\frac{S_t}{S_0}\right)}{\sigma}$, $\mu = \frac{\alpha-\frac{\sigma^2}{2}}{\sigma}$, $b^* = \frac{ln \left(\frac{B}{S_0}\right)}{\sigma}$, then $S_{t^*}$ hitting $B$ is equivalent to $X_{t^*}$ hitting $b^*$, where $X_t$ is a Brownian motion with drift.

The first passage time $\tau_{b^*}$ for $X_t$ hitting $b^*$ (i.e $t^*$ for $S_t$ hitting $B$) satisfies Inverse Gaussian: 
$$\tau_{b^*} \sim IG(\frac{b^*}{\mu}, {b^*}^2)$$

The Laplace transform of the first passage time $\tau_{b^*}$ for $X_t$ of bound $b^*$, i.e. $\tau_{b^*}$, is 
$$\mathbb E[e^{-r\tau_{b^*}}]=e^{b^*\mu-b^*\sqrt{2r+\mu^2}}, \forall r>0$$

- the expected discounted profit given $t$ is:
$$\mathbb E[v(t)] = \mathbb E[v(t)|\tau_{b^*}<t]F_{IG}(t)+\mathbb E[v(t)|\tau_{b^*}>t](1-F_{IG}(t)) $$

where
$$\mathbb E[v(t)|\tau_{b^*}<t] = B\left(\frac{B}{S_0} \right)^{\frac{\mu-\sqrt{2r_f+\mu^2}}{\sigma}} - S_0$$

$$\mathbb E[v(t)|\tau_{b^*}>t] = S_0\left(e^{(\alpha−r_f)t}-1\right) $$

- the variance of the discounted profit given $t$ is:

$$var(v(t))=\mathbb E[v^2(t)] - \mathbb E[v(t)]^2$$

- probability of loss given $t$ is:
$$\begin{split} P(v(t)<0)
& = P(\tau_{b^*} \geq \frac{\sigma m}{r_f}|\tau_{b^*}\leq t)P(\tau_{b^*}\leq t)+P(S_t<S_0 e^{r_f t}|\tau_{b^*}>t)P(\tau_{b^*}>t)
\end{split}$$

Without the third condition, buy and hold strategies yield statistical arbitrage opportunities in the Black-Scholes model for $\alpha−max(r_f,\frac{\sigma^2}{2})>0$ and $B>S_0$.

Proof: 

Consider the martingale $Z(t) = e^{\theta X(t) - (\theta \mu+\frac{\theta^2}{2})t}, \theta>0$ and $Z(0)=1$, there is 
$$\mathbb E[e^{\theta X(t\land \tau_{b^*}) - (\theta \mu+\frac{\theta^2}{2})(t\land \tau_{b^*})}]=1,t \geq 0$$
$$\lim_{t \to \infty} e^{\theta X(t\land \tau_{b^*}) - (\theta \mu+\frac{\theta^2}{2})(t\land \tau_{b^*})}=\mathbb 1_{\{\tau_{b^*}<\infty\}}e^{\theta b^* - (\theta \mu+\frac{\theta^2}{2}) \tau_{b^*}}$$

thus 
$$\mathbb E[\mathbb 1_{\{\tau_{b^*}<\infty\}}e^{\theta b^* - (\theta \mu+\frac{\theta^2}{2}) \tau_{b^*}}] = 1 $$
$$\mathbb E[\mathbb 1_{\{\tau_{b^*}<\infty\}} e^{-(\theta \mu+\frac{\theta^2}{2})\tau_{b^*}}] = \mathbb 1_{\{\tau_{b^*}<\infty\}} e^{-\theta b^*} $$
which holds for $(\theta \mu+\frac{\theta^2}{2})>0$ and $\theta>0$

- For the case $\mu<0$ and $\mu>\frac{-\theta}{2}$,$\theta$ can only converge to the positive constant, when $\theta \to -2\mu$, for which we obtain $$ P(\tau_{b^*} < \infty) = \mathbb E[\mathbb 1{\tau_{b^*} < \infty}] = e^{2\mu b^*}< 1,$$ for $\mu < 0$ and therefore $P(\tau_{b^*} < \infty) < 1$.

- For the case $\mu > 0$, we can take the limit on both sides for $\theta \to 0$, which yields $P(\tau_{b^*} <\infty)=1$, which means given $\sigma>0$ and $b^*>0$ (i.e. given $\alpha-\frac{\sigma^2}{2}>0$, $\sigma>0$, and $B>S_0$), $X_t$ hits the level $b^*$ almost surely with sufficiently large $t$. 

Let $\theta = \sigma$, then
$$\mathbb E[\mathbb 1_{\{\tau_{b^*}<\infty\}}e^{\sigma b^* - (\sigma \mu+\frac{\sigma^2}{2}) \tau_{b^*}}] = 1 $$

$$\begin{split} \mathbb E[e^{\sigma b^* - (\sigma \mu+\frac{\sigma^2}{2}) \tau_{b^*}}]
& = \mathbb E[\lim_{t \to \infty} \mathbb 1_{\{\tau_{b^*}<\infty\}}e^{\sigma b^* - (\sigma \mu+\frac{\sigma^2}{2}) \tau_{b^*}}] \\
& = \lim_{t \to \infty} \mathbb E[\mathbb 1_{\{\tau_{b^*}<\infty\}}e^{\sigma b^* - (\sigma \mu+\frac{\sigma^2}{2}) \tau_{b^*}}] \\
& = 1 \\
\end{split}$$

$$\mathbb E[e^{-\alpha \tau_{b^*}}] = e^{-m \alpha} = \frac{S_0}{B}$$

Then the expected discounted profit is 
$$\lim_{t \to \infty} \mathbb E[v(t)] = \lim_{t \to \infty} \mathbb E[v(t)|\tau_{b^*} \leq t] = B\left( \frac{B}{S_0} \right)^{\frac{\mu-\sqrt{2r_f+\mu^2}}{\sigma}} - S_0$$

for which if further $\alpha > r_f > 0$, there is $$\begin{split} \lim_{t \to \infty} \mathbb E[v(t)] 
& = \lim_{t \to \infty} \mathbb E[v(t)|\tau_{b^*} \leq t] \\
& = \lim_{t \to \infty} B \mathbb E[-e^{r_f \tau_{b^*}}] \\
& > \lim_{t \to \infty} B \mathbb E[-e^{\alpha \tau_{b^*}}] \\
& = B \frac{S_0}{B} - S_0 \\
& = 0 \\ 
\end{split}$$

And also there are

$$\lim_{t \to \infty} var(v(t)) = B^2 \left[\left(\frac{B}{S_0}\right)^{\frac{\mu-\sqrt{4r_f+\mu^2}}{\sigma}} - \left(\frac{B}{S_0}\right)^{\frac{2\mu - 2\sqrt{2r_f+\mu^2}}{\sigma}} \right]$$
$$\lim_{t \to \infty} \frac{var(v(t))}{t} = 0 $$ 

$$\text{ (violation) }\lim_{t \to \infty} P(v(t)<0) = P(\tau_{b^*} > \frac{\sigma b^*}{r_f}) = 1-F_{IG}(\frac{\sigma b^*}{r_f}) \neq 0$$


**Buy & Hold Until Deterministic Barrier Strategy**: a self-financing trading strategy with a deterministic boundary that consists of buying and holding one unit of stock financed by borrowing from the bank with constant risk-free rate. Whenever the stock price process hits to $S_0(1+k)e^{r_f t}$, sell the stock, realizing the profit of $k$, and invest immediately in the money market account.

$$v_t = \left\{
\begin{array} \
S_0 k, \text{ if } t^* \leq t & \\
S_te^{-r_f t} -S_0, \text{ else }, & \\
\end{array} \right. 
$$ where $t^* = min\left\{t \geq 0: S_t = S_0(1+k)e^{r_f t}\right\}$

Define $X_t = \frac{ln \left(\frac{S_t}{S_0}\right)}{\sigma}$, $\mu=\frac{\alpha-r_f-\frac{\sigma^2}{2}}{\sigma}$, $k^*=\frac{ln(1 + k)}{\sigma}$

The first passage time $t^*(\tau_{k^*}):\tau_{k^*} \sim IG(\frac{k^*}{\mu}, {k^*}^2)$.

- the expected discounted profit given $t$ is:
$$\mathbb E[v(t)] = \mathbb E[v(t)|\tau_{b^*}<t]F_{IG}(t)+\mathbb E[v(t)|\tau_{b^*}>t](1-F_{IG}(t)) $$

where
$$\mathbb E[v(t)|\tau_{b^*}<t] = k S_0$$

$$\mathbb E[v(t)|\tau_{b^*}>t] = S_0\left(e^{(\alpha−r_f)t}-1\right) $$

- the variance of the discounted profit given $t$ is:

$$var(v(t))=\mathbb E[v^2(t)] - \mathbb E[v(t)]^2$$

- probability of loss given $t$ is:
$$\begin{split} P(v(t)<0)
& = P(S_t < S_0 e^{r_f t},\tau_{k^∗} > t) \\
& = P(S_t < S_0 e^{r_f t}|\tau_{k^∗} > t)P(\tau_{k^∗} > t) \\
& = P((\alpha − r_f − \frac{\sigma^2}{2})t + \sigma W_t < 0|M(t) < k^∗)P(M(t) < k^∗) \\
& = P(Z < −\frac{(\alpha − r_f − \frac{\sigma^2}{2})\sqrt t}{\sigma}|\tau_{k^∗} > t)  P(\tau_{k^∗}>t) \\
& = P(Z < −\frac{(\alpha − r_f − \frac{\sigma^2}{2})\sqrt t}{\sigma}|\tau_{k^∗} > t) \times \left(\Phi(\frac{k^*-\frac{(\alpha−r_f-\frac{\sigma^2}{2})t}{\sigma}}{\sqrt t}) -e^{2k^*\frac{\alpha−r_f-\frac{\sigma^2}{2}}{\sigma}}\Phi(\frac{-k^*-\frac{(\alpha−r_f-\frac{\sigma^2}{2})t}{\sigma}}{\sqrt t})\right) \\
\end{split}$$
$$\text{where } M(t) \text{ is the maximum of the process} X_t \text{ in the time interval } [0,t],\; Z \text{ is the standard normal random variable, and } \Phi \text{ is the standard normal cdf}$$ 

For $\alpha−r_f \geq \frac{\sigma^2}{2}$ as $t \to \infty$ the probability of loss goes to zero, whereas for $\alpha−r_f < \frac{\sigma^2}{2}$ we almost surely make a loss as $t \to \infty$, the probability of loss does not decay to zero, with
$$\lim_{t \to \infty} P(v(t)<0) = 1-e^{2k^*\frac{\alpha−r_f-\frac{\sigma^2}{2}}{\sigma}} \text{ for } \alpha−r_f < \frac{\sigma^2}{2}$$

Statistical arbitrage opportunity exists when $\alpha−r_f \geq \frac{\sigma^2}{2}$ and $k > 0$.

Proof:

When $\mu > 0$ and $k^* > 0$ (i.e. $\alpha-r_f-\frac{\sigma^2}{2}>0$ given $\sigma>0$, and $k>0$), for sufficiently large t, $P(\tau_{k^*} < \infty)=1$, thus:

$$\lim_{t \to \infty} \mathbb E[v(t)] = \mathbb E [\lim_{t \to \infty} v(t)] = S_0k > 0$$
$$\lim_{t \to \infty} \frac{var(v(t))}{t} = 0 $$ 

$$\lim_{t \to \infty} P(v(t)<0) = 0 \text{ for } \alpha−r_f \geq \frac{\sigma^2}{2}$$


**Short Until Deterministic Barrier Strategy**:a self-financing trading strategy with a deterministic boundary that consists of shorting one unit of stock and investing in the money market account with constant risk-free rate. We close the short position whenever the stock price hits the boundary level, $S_0(1 + k)^{−1}e^{r_f t}$.

$$v_t = \left\{
\begin{array} \
\frac{S_0 k}{k+1}, \text{ if } t^* \leq t & \\
S_0 - S_te^{-r_f t}, \text{ else }, & \\
\end{array} \right. 
$$ where $t^* = min\left\{t \geq 0: S_t = S_0(1+k)^{-1}e^{r_f t}\right\}$

Define $\tilde X_t = -\frac{ln \left(\frac{S_t}{S_0}\right)}{\sigma}=-X_t$, $\tilde \mu=\frac{\frac{\sigma^2}{2}-(\alpha-r_f)}{\sigma}=-\mu$, $\tilde k^*=-\frac{ln(1 + k)}{\sigma}=-k^*$, then the conclusions of from the first passage time *Buy & Hold Until Deterministic Barrier Strategy* still apply

The first passage time $t^*(\tau_{k^*}):\tau_{k^*} \sim IG(\frac{k^*}{\mu}, {k^*}^2)$.

- the expected discounted profit given $t$ is:
$$\mathbb E[v(t)] = \mathbb E[v(t)|\tau_{b^*}<t]F_{IG}(t)+\mathbb E[v(t)|\tau_{b^*}>t](1-F_{IG}(t)) $$

where
$$\mathbb E[v(t)|\tau_{b^*}<t] = \frac{S_0 k}{k+1}$$

$$\mathbb E[v(t)|\tau_{b^*}>t] = -S_0\left(e^{(\alpha−r_f)t}-1\right) $$

- the variance of the discounted profit given $t$ is:

$$var(v(t))=\mathbb E[v^2(t)] - \mathbb E[v(t)]^2$$

- probability of loss given $t$ is:
$$\begin{split} P(v(t)<0)
& = P(S_t > S_0 e^{r_f t},\tau_{k^∗} > t) \\
& = P(S_t > S_0 e^{r_f t}|\tau_{k^∗} > t)P(\tau_{k^∗} > t) \\
& = P((\alpha − r_f − \frac{\sigma^2}{2})t + \sigma W_t > 0|m(t) > -k^∗)P(m(t) > -k^∗) \\
& = P(Z > −\frac{(\alpha − r_f − \frac{\sigma^2}{2})\sqrt t}{\sigma}|\tau_{k^∗} > t)  P(\tau_{k^∗}>t) \\
& = P(Z < \frac{(\alpha − r_f − \frac{\sigma^2}{2})\sqrt t}{\sigma}|\tau_{k^∗} > t) \times \left(\Phi(\frac{k^*-\frac{(\alpha−r_f-\frac{\sigma^2}{2})t}{\sigma}}{\sqrt t}) -e^{2k^*\frac{\alpha−r_f-\frac{\sigma^2}{2}}{\sigma}}\Phi(\frac{-k^*-\frac{(\alpha−r_f-\frac{\sigma^2}{2})t}{\sigma}}{\sqrt t})\right) \\
\end{split}$$
$$\text{where } M(t) \text{ is the maximum of the process} X_t \text{ in the time interval } [0,t],\; Z \text{ is the standard normal random variable, and } \Phi \text{ is the standard normal cdf}$$ 

Statistical arbitrage opportunity exists when $\alpha−max(r_f, \frac{\sigma^2}{2}) < 0$ and $k > 0$.

Proof:

When $\tilde \mu > 0$ and $\tilde k^* > 0$(i.e. $\frac{\sigma^2}{2}-(\alpha-r_f)>0$ given $\sigma>0$, and $k>0$), for sufficiently large t, $P(\tau_{\tilde k^*} < \infty)=1$, thus:

$$\lim_{t \to \infty} \mathbb E[v(t)] = \mathbb E [\lim_{t \to \infty} v(t)|\tau_{b^*}<t] = \frac{S_0 k}{k+1} > 0$$
$$\lim_{t \to \infty} \frac{var(v(t))}{t} = 0 $$ 
$$\lim_{t \to \infty} P(v(t)<0) = 0 \times (1-e^{2k^*\frac{\alpha−r_f-\frac{\sigma^2}{2}}{\sigma}}) = 0 \text{ for } \alpha−r_f < \frac{\sigma^2}{2}$$
